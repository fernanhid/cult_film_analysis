```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
final_file = pd.read_pickle('movies_df_2005-2015.pkl')
```

```python
# Get rid of missing Values
cleaner_df = final_file[(final_file.new_gross != 'NaN')&\
                (final_file.ratings != 'no rating')&\
                (final_file.runtime_in_minutes != 'NaN')&\
                (final_file.new_opening_weekend != 'Nan')]
```

```python
plt.figure().set_size_inches([10,6])
plt.scatter(y = 'ratings', x = 'new_gross', data = cleaner_df)
plt.xlabel('Domestic Gross')
_ = plt.ylabel('Ratings')
```

```python
cult_df = cleaner_df[(cleaner_df.ratings > 85)&\
                     (cleaner_df.new_gross < 70000000)&\
                     (cleaner_df.movie_titles !='The Wizard of Oz (1939) - IMDb')]

ax = cult_df.plot(x = 'new_gross',
                     y = 'ratings',
                     kind = 'scatter',
                     figsize = (20,10),
                     color = cult_df.ratings,
                     colormap = 'summer',
                     colorbar= False,
                     fontsize = 20,
                     s = 300)

ax.set_xlabel('Gross', fontsize=20)
ax.set_ylabel('Ratings', fontsize=20)
ax.set_title('Cult Movies', fontsize = 30)


for label, x, y in zip(cult_df.movie_titles, 
                       cult_df.new_gross,
                       cult_df.ratings):
    ax.annotate(label, xy = (x, y)  , xytext = (-10, 20), size = 10,
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        arrowprops = dict(arrowstyle = '-', connectionstyle = 'arc3,rad=0'))
 
```

```python
anti_cult_df = cleaner_df[(cleaner_df.ratings < 30)&\
                          (cleaner_df.new_gross > 40000000)&\
                          (cleaner_df.movie_titles !='The Wizard of Oz (1939) - IMDb')]

ax = anti_cult_df.plot(x = 'new_gross',
                     y = 'ratings',
                     kind = 'scatter',
                     figsize = (20,10),
                     color = anti_cult_df.ratings,
                     colormap = 'summer',
                     colorbar= False,
                     fontsize = 20,
                     s = 300)

ax.set_xlabel('Gross', fontsize=20)
ax.set_ylabel('Ratings', fontsize=20)
ax.set_title('Anti-Cult Movies', fontsize = 30)


for label, x, y in zip(anti_cult_df.movie_titles, 
                       anti_cult_df.new_gross,
                       anti_cult_df.ratings):
    ax.annotate(label, xy = (x, y)  , xytext = (-10, 20), size = 10,
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        arrowprops = dict(arrowstyle = '-', connectionstyle = 'arc3,rad=0'))
```
